```python
def Frequentie(T, I, O):
    n = len(T)
    AantalperiodesI = 0
    for i in range(n-1):
        if I[i] > 0 and I[i+1] < 0:
            AantalperiodesI = AantalperiodesI + 1
        if I[i] < 0 and I[i+1] > 0:
            AantalperiodesI = AantalperiodesI + 1
    if O.max()-O.min() > 0.2 :
        AantalperiodesO = 0
        for i in range(n-1):
            if O[i] > 0 and O[i+1] < 0:
                AantalperiodesO = AantalperiodesO + 1
            if O[i] < 0 and O[i+1] > 0:
                AantalperiodesO = AantalperiodesO + 1
        AantalPeriodes = (AantalperiodesI+AantalperiodesO)/4
    else:
        AantalPeriodes = AantalperiodesI/2
    print(AantalPeriodes/5)
    return(AantalPeriodes/5)
```

```python
def AmpModel(f, Amplitude, RC):
    #print(Amplitude, RC)
    teller = 2*np.pi* Amplitude * RC * f
    noemer = np.sqrt(1 + (RC * 2*np.pi* f)**2)
    return(teller/noemer)

def Amp(f, Amplitude, RC):
    opl = []
    for i in f:
        opl.append(AmpModel(i, Amplitude, RC))
    return(opl)
```
